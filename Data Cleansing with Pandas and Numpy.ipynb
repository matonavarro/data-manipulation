```python
#Let's read the full csv data set and display the top records
import pandas as pd
data = pd.read_csv('../reading files/Customer-Churn.csv')
print(data.head())
```

```python
#Let's check for nulls and view the data type on each of the columns
data.info()
```

```python
#Lets convert TotalCharges to a numeric data type
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
```

```python
#lets find all records that have an empty string value in TotalCharges
data.loc[data['TotalCharges'] == ' ']
```

```python
#Let's replace ' ' with 0 and validate
data['TotalCharges'] = data['TotalCharges'].replace(' ', 0)
data.loc[data['TotalCharges'] == ' ']
```

```python
#Let's replace 0 in TotalCharges with the MonthlyCharges value
import numpy as np
data['TotalCharges'] = np.where(data['TotalCharges']== 0, data['MonthlyCharges'],data['TotalCharges'])
```

```python
#Let's try converting Total Charges to numeric again
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
data.info()
```

```python
# Let's replace 'No Internet Service' values with 'No'
replace_cols = ['OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
for i in replace_cols:
    data[i] = data[i].replace({'No internet service':'No'})
```

```python
#Let's replace 'No Phone Service' with No
data['MultipleLines'] = data['MultipleLines'].replace({'No phone service':'No'})
```

```python
#Let's replace Senior Citizen values from 1 and 0 to yes and no
data['SeniorCitizen'] = data['SeniorCitizen'].replace({1:'Yes', 0:'No'})
```

```python
#Let's create Tenure groups
def grouptenure(data):
    
    if data['tenure']<= 10:
        return '10 or less'
    elif (data['tenure']>10) & (data['tenure']<=20):
        return '11-20'
    elif (data['tenure']>20) & (data['tenure']<=30):
        return '21-30'
    elif (data['tenure']>30) & (data['tenure']<=40):
        return '31-40'
    elif (data['tenure']>40) & (data['tenure']<=50):
        return '41-50'
    elif (data['tenure']>50) & (data['tenure']<=60):
        return '51-60'
    elif (data['tenure']>60):
        return 'greater than 60'
data['TenureGroup'] = data.apply(lambda data:grouptenure(data), axis=1)
```

### Data Cleansing Summary
> -  The Total Charges column (which should be a number) is stored as an object.  To perform numerical analysis on the Total Charges column we converted the data type from object to float64.
-  During the Total Charges conversion, we found 11 records where the Total Charge value was an empty string.  Business context about the data allows us to replace these empty strings with a value of 0.
-  Six columns revealed 3 values which could be combined to 2, where 'No Internet Service' or 'No Phone Service'  was converted to 'No'.
-  For consistency, we have replaced the 1 and 0 values in SeniorCitizen with 'Yes' and 'No'.   
-  The Tenure colum contains a wide range of values, which may introduce overfitting issues during modelling, therefore we created a higher level of tenure order by creating Tenure buckets.
